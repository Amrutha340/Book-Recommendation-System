```python
users = pd.read_csv('/content/Book reviews/BX-Users.csv', sep=';',encoding="latin-1")
users.columns = ['userID', 'Location', 'Age']
ratings = pd.read_csv('/content/Book reviews/BX-Book-Ratings.csv', sep=';',  encoding="latin-1")
ratings.columns = ['userID', 'ISBN', 'bookRating']
books = pd.read_csv('/content/Book reviews/BX_Books.csv', sep=';',encoding="latin-1")
books.columns = ['ISBN', 'bookTitle', 'bookAuthor', 'yearOfPublication', 'publisher', 'imageUrlS', 'imageUrlM', 'imageUrlL']
```

# Shape of Datasets

```python
print("Users -",users.shape)
print("Books -",books.shape)
print("Ratings -",ratings.shape)
```

# Books

```python
books.head(5)
```

Dropping columns that are not required from books dataset

```python
books.drop(['imageUrlS', 'imageUrlL'],axis = 1,inplace = True)
books.head(5)
```

Datatypes

```python
books.dtypes
```

```python
books.describe(include ='all')
```

yearOfPublication

```python
print("Before :",sorted(books['yearOfPublication'].unique()))
```

```python
#this dataset was published in 2004
#setting invalid years as NaN
books.loc[(books.yearOfPublication > 2004) | (books.yearOfPublication < 1800),'yearOfPublication'] = np.NAN
#replacing NaNs with mean value of yearOfPublication
books.yearOfPublication.fillna(round(books.yearOfPublication.mean()), inplace=True)
print("After :",sorted(books['yearOfPublication'].unique()))
```

```python
books.yearOfPublication = books.yearOfPublication.astype(np.int32)
```

```python
books.isnull().sum()
```

```python
books.bookAuthor = books.bookAuthor.fillna('Other')
books.publisher = books.publisher.fillna('Other')
```

# Users

```python
users.head(5)
```

```python
Location = users.Location.str.split(', ', n=2, expand=True)
users['City'] = Location[0]
users['State'] = Location[1]
users['Country'] = Location[2]
users.drop('Location',axis=1,inplace = True)
```

```python
users.dtypes
```

```python
users.describe(include = 'all')
```

Age

```python
print(sorted(users.Age.unique()))
```

```python
#below 5 and above 90 dont make sense
#hence replacing these by NaNs
users.loc[(users.Age > 90) | (users.Age < 5), 'Age'] = np.nan
#replacing NaNs with mean
users.Age = users.Age.fillna(users.Age.mean())
#setting the data type as int
users.Age = users.Age.astype(np.int32)
#rechecking
print(sorted(users.Age.unique()))
```

```python
users.isnull().sum()
```

```python
users.State = users.State.fillna('Other')
users.Country = users.Country.fillna('Other')
```

# Ratings

```python
ratings.head(5)
```

```python
ratings.dtypes
```

```python
ratings.describe(include='all')
```

```python
ratings.isnull().sum()
```

```python
#ratings dataset should have books only which exist in our books dataset
ratings = ratings[ratings.ISBN.isin(books.ISBN)]
#ratings dataset should have users only which exist in our users dataset
ratings = ratings[ratings.userID.isin(users.userID)]
```

Removing implicit ratings

```python
print(sorted(ratings.bookRating.unique()))
#Ratings are either explicit, expressed on a scale from 1-10 
#or implicit, expressed by 0
```

```python
ratings = ratings[ratings.bookRating != 0]
print(sorted(ratings.bookRating.unique()))
print("Shape :",ratings.shape)
```

Reducing Dimensionality

```python
#Remove items which were rated less than 10 times
a = ratings.groupby('ISBN').filter(lambda x: len(x) >= 10)
#Remove users who gave less than 15 ratings
ratings = a.groupby('userID').filter(lambda x: len(x) >= 15)
```

```python
print("Ratings shape :",ratings.shape)
print("No of unique users :",len(np.unique(ratings['userID'])))
print("No of unique books :",len(np.unique(ratings['ISBN'])))
```
